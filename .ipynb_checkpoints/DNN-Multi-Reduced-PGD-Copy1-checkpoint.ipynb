```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import shap
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from sklearn.manifold import TSNE
from sklearn.metrics import  accuracy_score, classification_report, confusion_matrix, f1_score, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from torch.utils.data import TensorDataset, DataLoader
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
df = pd.read_csv("undersampled_CIC2019_dataset.csv")
# df = pd.read_csv("/content/DIS_Hughen/undersampled_CIC2019_dataset.csv")
```

```python
df.shape
```

```python
df["label"].value_counts()
```

```python
dos_set = df[df["label"].str.contains("DoS", case=False, na=False)]
dos_sample = (dos_set.sample(n=50000, random_state=42, replace=False).reset_index(drop=True))
dos_sample["label"] = "DoS_ATTACK"
dos_sample["label"].value_counts()
```

```python
non_dos_set = df[(df["label"] != "BENIGN") & (~df["label"].str.contains("DoS", case=False, na=False))]
non_dos_sample = (non_dos_set.sample(n=50000, random_state=42, replace=False).reset_index(drop=True))
non_dos_sample["label"] = "NON_DoS_ATTACK"
non_dos_sample["label"].value_counts()
```

```python
attack_sample =pd.concat([dos_sample, non_dos_sample]).reset_index(drop=True)
attack_sample["label"].value_counts()
```

```python
benign_set = df[df["label"] == "BENIGN"]
benign_sample = benign_set.sample(n=non_dos_sample.shape[0], replace=False, random_state=42)
```

```python
benign_sample.shape
```

```python
print(benign_sample["label"].value_counts())
```

```python
df_sample =pd.concat([benign_sample, attack_sample]).reset_index(drop=True)
```

```python
df_sample.shape
```

```python
df_sample["label"].value_counts()
```

```python
le = LabelEncoder()
df_sample["label"] = le.fit_transform(df_sample["label"])
X = df_sample.drop(columns=["label"], errors="ignore").select_dtypes(include=[np.number])
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.fillna(0, inplace=True)
y = df_sample["label"]
label_mapping = dict(zip(le.classes_, range(len(le.classes_))))
print(label_mapping)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y # 0.2
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.long)

X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.values, dtype=torch.long)

train_loader = DataLoader(TensorDataset(X_train_tensor, y_train_tensor), batch_size=2048, shuffle=True)
val_loader = DataLoader(TensorDataset(X_val_tensor, y_val_tensor), batch_size=2048)
```

```python
input_size = 88
hidden_size = 128
num_classes = 3

class MultiClassDNN(nn.Module):
    def __init__(self, input_size=88, num_classes=3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_size, 128), #256
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, num_classes)
        )
    def forward(self, x):
        return self.net(x)
```

```python
model = MultiClassDNN().to(device)
criterion = nn.CrossEntropyLoss(label_smoothing=0.05)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=0.0001)
```

```python
num_epochs = 100
train_losses = []
val_losses = []
for epoch in range(num_epochs):
    model.train()
    train_loss = 0
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        logits = model(X_batch)
        loss = criterion(logits, y_batch)
        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        train_loss += loss.item() * X_batch.size(0)
    avg_train_loss = train_loss / len(train_loader.dataset)
    train_losses.append(avg_train_loss)
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            logits = model(X_batch)
            loss = criterion(logits, y_batch)
            val_loss += loss.item() * X_batch.size(0)
    avg_val_loss = val_loss / len(val_loader.dataset)
    val_losses.append(avg_val_loss)
    print(f"Epoch {epoch+1}, Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")
    
torch.save(model.state_dict(), "multiclass_dnn.pth")
```

```python
plt.figure()
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.grid()
plt.show()
```

```python
model.eval()
X_tensor = X_val_tensor.to(device)
y = y_val.values
with torch.no_grad():
    logits = model(X_tensor)
    y_prob = torch.softmax(logits, dim=1).cpu().numpy()
    y_pred = np.argmax(y_prob, axis=1)
```

```python
print("Accuracy:", accuracy_score(y, y_pred))
print("Weighted F1:", f1_score(y, y_pred, average='weighted'))
print("Macro F1:", f1_score(y, y_pred, average='macro'))
print("ROC-AUC:", roc_auc_score(y, y_prob, multi_class="ovr", average="weighted"))
```

```python
print(classification_report(y, y_pred, target_names=le.classes_))
```

```python
cont_idx = [0, 1, 2, 3, 4]
bin_idx = [5, 6, 7]
cat_groups = [[8, 9, 10], [11, 12, 13]]

def pgd_attack(model, x, y, epsilon = 0.1, alpha = 0.01, num_steps = 200, n_restarts = 5, momentum = 0.9, x_min = None, x_max = None, target_class = None, verbose = False):
    model.eval()
    device = next(model.parameters()).device
    if not isinstance(x, torch.Tensor):
        x = torch.tensor(x, dtype=torch.float32)
    if not isinstance(y, torch.Tensor):
        y = torch.tensor(np.array(y), dtype=torch.long)
    x = x.to(device)
    y = y.to(device)
    cont_idx_t = torch.tensor(cont_idx, device=device)
    bin_idx_t  = torch.tensor(bin_idx,  device=device) if bin_idx else None  
    best_loss = torch.full((x.size(0),), -float("inf"), device=device)
    best_x_adv = x.clone()
    for restart in range(n_restarts):
        delta = torch.zeros_like(x)
        delta[:, cont_idx] = torch.empty(x.size(0), len(cont_idx), device=device).uniform_(-epsilon, epsilon)
        x_adv = (x + delta).clone()
        if x_min is not None and x_max is not None:
            x_adv = torch.clamp(x_adv, x_min, x_max)
        grad_momentum = torch.zeros_like(x)
        for step in range(num_steps):
            x_adv = x_adv.detach().requires_grad_(True)
            logits = model(x_adv)
            if target_class is not None:
                # Targeted: minimise loss for target → negate so we maximise it
                t = torch.full_like(y, target_class)
                loss = -F.cross_entropy(logits, t)
            else:
                loss = F.cross_entropy(logits, y)
            model.zero_grad(set_to_none=True)
            loss.backward()
            raw_grad = x_adv.grad.detach()
            l1_norm = raw_grad.abs().sum(dim=1, keepdim=True).clamp(min=1e-12)
            grad_momentum = momentum * grad_momentum + raw_grad / l1_norm
            grad_sign = torch.sign(grad_momentum)
            x_adv = x_adv.detach()
            x_adv[:, cont_idx_t] = x_adv[:, cont_idx_t] + alpha * grad_sign[:, cont_idx_t]
            # Project back into ε-ball around original x
            x_adv[:, cont_idx_t] = torch.max(
                torch.min(x_adv[:, cont_idx_t], x[:, cont_idx_t] + epsilon),
                x[:, cont_idx_t] - epsilon,
            )
            if bin_idx_t is not None and len(bin_idx) > 0:
                flip_mask = (grad_sign[:, bin_idx_t] > 0)   # flip toward 1
                x_adv[:, bin_idx_t] = torch.where(flip_mask,
                    torch.ones_like(x_adv[:, bin_idx_t]),
                    torch.zeros_like(x_adv[:, bin_idx_t]),
                )
            if cat_groups is not None:
                for group in cat_groups:
                    g = torch.tensor(group, device=device)
                    # Start from current one-hot, nudge by gradient, re-project
                    logits_cat = x_adv[:, g] + alpha * grad_sign[:, g]
                    best_cat   = torch.argmax(logits_cat, dim=1)
                    one_hot    = torch.zeros_like(logits_cat)
                    one_hot[torch.arange(x.size(0)), best_cat] = 1.0
                    x_adv[:, g] = one_hot
            if x_min is not None and x_max is not None:
                x_adv = torch.clamp(x_adv, x_min, x_max)
        with torch.no_grad():
            logits_final = model(x_adv)
            if target_class is not None:
                t = torch.full_like(y, target_class)
                per_sample_loss = -F.cross_entropy(logits_final, t, reduction="none")
            else:
                per_sample_loss = F.cross_entropy(logits_final, y, reduction="none")
        improved = per_sample_loss > best_loss
        best_loss  = torch.where(improved, per_sample_loss, best_loss)
        best_x_adv = torch.where(improved.unsqueeze(1).expand_as(x_adv), x_adv, best_x_adv)
        if verbose:
            fool_rate = (torch.argmax(logits_final, dim=1) != y).float().mean().item()
            print(f"  Restart {restart+1}/{n_restarts} | loss={per_sample_loss.mean():.4f} | fool_rate={fool_rate:.3f}")
    return best_x_adv.detach()
```

```python
def epsilon_eval(model, X_val, y_val, epsilon, le=None, n_restarts=5, momentum=0.9, verbose=True):
    x_adv = pgd_attack(model, X_val, y_val, epsilon=epsilon)
    device = next(model.parameters()).device
    model.eval()
    with torch.no_grad():
        logits   = model(x_adv.to(device))
        y_pred   = torch.argmax(logits, dim=1).cpu().numpy()
    y_true = y_val if isinstance(y_val, np.ndarray) else np.array(y_val)
    from sklearn.metrics import accuracy_score, classification_report
    print("=== ADVERSARIAL EVALUATION ===")
    print(f"Epsilon : {epsilon}")
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    tgt = le.classes_ if le is not None else None
    print(classification_report(y_true, y_pred, target_names=tgt))

for epsilon in [0.01, 0.05, 0.1, 0.5]:
    epsilon_eval(model, X_val, y_val, epsilon, le=le, n_restarts=5, momentum=0.9, verbose=True)
```
